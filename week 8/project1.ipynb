```python
import tkinter as tk
from tkinter import messagebox

class Zenith:
    def unique_services(self):
        pass  # Will be overridden

    def mutual_services(self):
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance"
        ]

class RetailBanking(Zenith):
    def unique_services(self):
        return [
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]


class GlobalBanking(Zenith):
    def unique_services(self):
        return [
            "Multi-currency management services",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]


class CommercialBanking(Zenith):
    def unique_services(self):
        return [
            "Advisory services"
        ]

def show_services():
    name = name_entry.get()
    division = division_var.get()

    if not name or not division:
        messagebox.showerror("Error", "Please enter name and select division")
        return

    
    if division == "Retail":
        bank = RetailBanking()
    elif division == "Global":
        bank = GlobalBanking()
    elif division == "Commercial":
        bank = CommercialBanking()
    else:
        messagebox.showerror("Error", "Invalid division")
        return

    unique = bank.unique_services()
    mutual = bank.mutual_services()

    result = f"Employee: {name}\nDivision: {division}\n\n"
    result += "Unique Services:\n" + "\n".join(unique) + "\n\n"
    result += "Mutual Services:\n" + "\n".join(mutual)

    result_label.config(text=result)


root = tk.Tk()
root.title("Zenith Bank Service System")
root.geometry("500x500")

# Name Input
tk.Label(root, text="Enter Employee Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack(pady=5)

tk.Label(root, text="Select Division:").pack(pady=5)
division_var = tk.StringVar()

tk.Radiobutton(root, text="Retail Banking", variable=division_var, value="Retail").pack()
tk.Radiobutton(root, text="Global Banking", variable=division_var, value="Global").pack()
tk.Radiobutton(root, text="Commercial Banking", variable=division_var, value="Commercial").pack()

tk.Button(root, text="Show Services", command=show_services).pack(pady=10)

result_label = tk.Label(root, text="", justify="left")
result_label.pack(pady=10)
root.mainloop()
```

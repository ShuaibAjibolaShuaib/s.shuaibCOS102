```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Andre",
            "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

    def check_employee(self, name):
        return name in self.employees

    def take_attendance(self, name):
        return f"Attendance taken for {name}"

    def assign_task(self):
        return random.choice(self.tasks)

    def refuse_access(self):
        return "Access Denied! You are not an employee."


# GUI PART
emp = Employee()

def submit():
    name = entry.get()

    if emp.check_employee(name):
        attendance = emp.take_attendance(name)
        task = emp.assign_task()

        messagebox.showinfo(
            "Success",
            f"{attendance}\nAssigned Task: {task}"
        )
    else:
        messagebox.showerror(
            "Error",
            emp.refuse_access()
        )


# Create window
root = tk.Tk()
root.title("Employee System")
root.geometry("350x200")

# Label
label = tk.Label(root, text="Enter Employee Name:")
label.pack(pady=10)

# Entry
entry = tk.Entry(root, width=30)
entry.pack(pady=5)

# Button
btn = tk.Button(root, text="Submit", command=submit)
btn.pack(pady=20)

# Run app
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class Delivery:
    def calculate_fee(self, location, weight):
        if location == "PAU":
            if weight >= 10:
                return 2000
            else:
                return 1500

        elif location == "Epe":
            if weight >= 10:
                return 5000
            else:
                return 4000
        else:
            return None


# Create object
delivery = Delivery()

def calculate():
    location = location_var.get()
    
    try:
        weight = float(weight_entry.get())
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number for weight.")
        return

    fee = delivery.calculate_fee(location, weight)

    if fee is None:
        messagebox.showerror("Error", "Invalid location selected.")
    else:
        messagebox.showinfo(
            "Delivery Fee",
            f"Delivery to {location}\nWeight: {weight}kg\nAmount to pay: ₦{fee}"
        )

root = tk.Tk()
root.title("Delivery Charge System")
root.geometry("350x250")

# Location
tk.Label(root, text="Select Location:").pack(pady=5)

location_var = tk.StringVar()
location_menu = tk.OptionMenu(root, location_var, "PAU", "Epe")
location_menu.pack(pady=5)

# Weight
tk.Label(root, text="Enter Package Weight (kg):").pack(pady=5)

weight_entry = tk.Entry(root)
weight_entry.pack(pady=5)

# Button
tk.Button(root, text="Calculate Fee", command=calculate).pack(pady=20)

# Run app
root.mainloop()
```
